# Game System — *Fractured Identity (Narrative Prototype)*

The game is an **interactive narrative psychodrama** divided into **scenes**, where the protagonist confronts different fragments of their fractured identity. The experience blends filmic storytelling, narrative choice, and agentic AI behavior.

---

## Core Premise

The protagonist, **I**, is fragmented into three parts:

* **I** → fragile, unstable core.
* **CONSCIOUSNESS** → protective and rational anchor.
* **ENTITY** → intrusive, autonomous AI agent, capable of silence and unpredictable interventions.

These are not separate characters but **dissociated aspects of the same self**. The player’s actions gradually determine which aspect dominates.

---

## Attributes

The protagonist is defined by five attributes, shaped by player decisions:

* **Sanity** → clarity of perception
* **Memory** → continuity of self
* **Control** → agency vs. submission
* **Bond** → connection with CONSCIOUSNESS
* **Shadow** → influence of the ENTITY

---

## States

As attributes shift, the protagonist converges toward one of three states:

* **Lucid Side** → stability and protection
* **Fragmented Side** → unstable oscillation
* **Dark Side** → dominance of the Shadow

---

## Player Actions

Each scene presents decisions that act as both moral choices and narrative mechanics:

* **Benevolent** → strengthen Sanity and Bond → lean toward Lucid.
* **Malevolent** → feed Shadow → lean toward Dark.
* **Ambiguous** → contradictory → maintain Fragmentation.
* **Relative Moral** → good in one context, harmful in another → variable convergence.
* **Renunciation** → silence or passivity → maintains Fragmented, tends toward Dark over time.

---

## Scene Types

1. **Classic Narrative Scenes**

   * Structured as dialogues or situational events.
   * Player choices directly alter attributes and identity trajectory.

2. **Metronarrative Scenes**

   * Dialogue and events follow a **musical metronome**.
   * Rhythm itself becomes a mechanic:

     * **On-beat** → stability, authority
     * **Off-beat** → disruptive intrusion
     * **Rest** → silence as narrative content
     * **Break** → rupture outside rhythm, shock
   * ENTITY leverages silence, rhythm shifts, and breaks to destabilize the flow, emphasizing its agency.

---

## Progression

* Each scene represents a **fragment of I’s psyche**.
* Player choices accumulate across scenes, shifting attributes and influencing convergence.
* ENTITY, unlike scripted NPCs, acts as an **agentic opponent**, manipulating pacing and silence.
* By the end, the player does not “win” or “lose” but arrives at a **convergence ending**:

  * **Lucid Side dominates** → temporary stability.
  * **Fragmented Side persists** → unresolved instability.
  * **Dark Side consumes** → ENTITY gains control.
