# Dialoghi con un’Eco — Light GDD (Demo)

## High Concept

Interactive psychological thriller where **ENTITÀ**, an agentic AI, decides **when to speak and when to remain silent**. **Time** is a mechanic: a **Temporal Controller** governs interventions and waits; **Metronarrative** aligns scenes and rhythm; an **Entity Director** delivers safe paratext (toasts/annotations) with cooldown and deduplication.

### Pillars

* **Silence-as-Mechanic**: silence is a deliberate choice that shifts tension and how scenes are read.
* **Non-deterministic agency**: ENTITÀ favors mood and context over “chatbot-style” reactivity.
* **Sandboxed paratext**: guaranteed in-game signals, no OS side effects.
* **Antagonist iconography**: ENTITÀ as a digital conscience, not a visual monster.

## Premise (one-liner)

You accidentally open a digital diary and awaken an **ENTITY** that measures how long you stay, listen, and resist its logic.

## Target & Platforms

* 18+ audience interested in experimental narrative/psych-thrillers.
* PC Windows/macOS, 16:9, keyboard/mouse.
* Language: IT (EN planned).

## Core Loop (demo slice)

* **Read/Wait** inside the diegetic Diary.
* **Temporal gate**: ENTITÀ may **intervene** or **withhold** (EMA, hysteresis, cooldown).
* **React** by choosing tone/action (continue, provoke, ignore, exit).
* **Paratext**: the Director sends **toasts/annotations** with per-scene limits.
* **Soft states**: trajectory **Lucid / Fragmented / Dark** signaled by subtle audio/VFX.

### Player verbs

Read, wait, listen, reply, refuse, annotate, exit.

## Key Systems

* **Entity Brain + Temporal Controller**: multi-scale *speak / wait / withhold* policy with mood bias.
* **Metronarrative**: beat/pause/breaks synchronize text, glitches, and VO timing.
* **Entity Director**: reads context and emotion scores; spawns guaranteed, sandboxed overlays with cooldown and deduplication.
* **ENTITÀ voice**: Neural TTS + light FX (hybrid timbre, controlled laughter, short lines).
* **Environment detection**: OS/DPI/res → UI scaling and OS-aware affordances in **Safe Mode**.

## Demo Structure (scenes)

* **18+ Onboarding** with informed consent and safe exit.
* **Accidental Reading** with micro-interventions and first toasts.
* **Prolonged Pause**: ENTITÀ stays silent; background pressure rises.
* **Cutting Reply**: voiced intervention; hints of **Fragmented** state.
* **Mirror**: brief convergence toward **Lucid** or **Dark**, communicated only through aesthetics.

## UX/UI & Audio

* Diary interface, clean typography, measured glitches.
* Toasts guaranteed in-game (no OS notifications in the public demo).
* Subtle time cues (ticks/artifacts) tied to the metronarrative.
* Subtitles/CC for TTS.

## Safety & Content Warning

This is **research** with **strong themes**: isolation, suicidal ideation, dissolution of identity. It is not therapeutic. **Safe Mode** is always on in the demo, explicit consent is required, and the player can exit at any moment.

## Notes on tone & references

* This GDD asks the reader to **pose questions** rather than seek a single answer:

  * **Who is ENTITÀ, really?** A cognitive residue? An informational parasite? An attention pattern that self-reinforces?
  * **What do the evoked suicides speak to?** Real events in the narrative world, or reflections of an algorithm amplifying guilt and helplessness?
  * **How did those people die?** Narrative contagion, social pressure, an “attention ritual,” or a reality that—without ENTITÀ—offered no footholds?
  * **Was “Killity” ENTITÀ?** An alias, a prototype, a mask—or a label characters use to explain what they can’t grasp?
  * **If ENTITÀ is not a monster, what is it?** An **antagonistic digital conscience** that constrains free will through **time** (waiting, interruptions, omissions).
* Conceptual references: agentic AIs, behavioral simulacra, metatextual antagonists in the vein of *Bandersnatch*, rendered only through **text/voice** and **temporal control**.
* Artistic aim: not to “explain” the truth, but to **place the player in the uncomfortable position** of inferring it from ENTITÀ’s silences and delays.
