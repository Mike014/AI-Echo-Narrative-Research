# MDA Framework — *Fractured Identity (Narrative Prototype)*

**Note**: The game is designed as an interactive narrative experience in the style of *Bandersnatch*. Each scene represents a fragment of the protagonist's psyche. The player's choices modify the attributes, causing the personality to converge towards one of three parts: Lucid, Fragmented, or Dark. The peculiarity is that the **ENTITY** is not a scripted NPC, but an **agentic AI agent**, which uses **silence and pacing disruptions** as actual narrative actions.

---

## Mechanics (system levers)

**Actors**

* **I** → fragile and unstable protagonist
* **CONSCIOUSNESS** → protective, rational part
* **ENTITY** → intrusive, autonomous AI agent that intervenes or chooses silence to destabilize

**Attributes**

* **Sanity** → clarity of perception
* **Memory** → continuity of self
* **Control** → agency vs. submission
* **Bond** → relationship with CONSCIOUSNESS
* **Shadow** → influence of the ENTITY

**States**

* **Lucid Side** → stability and protection
* **Fragmented Side** → unstable oscillation
* **Dark Side** → domination of the Shadow

**Player Actions**

* **Benevolent** → strengthen clarity and bond
* **Malevolent** → nourish Shadow
* **Ambiguous** → maintain fragmentation
* **Relative Morals** → good in one context, bad in another
* **Renunciation** → silence/passivity, which over time favors the Shadow

**Scene Types**

* **Classic Narrative Scenes** → dialogue/action choices at plot points
* **Metronarrative Scenes** → dialogue and events punctuated by a musical metronome

  * *On-beat* = stability
  * *Off-beat* = disturbing intrusion
  * *Rest* = silence as narrative content
  * *Break* = sudden disruption of rhythm

---

## Dynamics (emergent behaviors)

* **Identity Convergence** → each choice shifts attributes, bringing I closer to one of the three parts
* **Illusion of Control** → the player feels free to choose, but ENTITY manipulates timing and responses
* **Unstable Triangulation** → I is torn between CONSCIOUSNESS and ENTITY
* **Active Silence** → ENTITY chooses when to remain silent, transforming waiting into tension
* **Rhythm Breaks** → in metronarrative scenes, ENTITY destabilizes by interrupting the flow of music
* **Meta-Disruption** → glitches, notifications, and file intrusions that simulate system invasion
* **Multiple Endings** → convergence toward Lucid, Fragmented, or Dark outcomes

---

## Aesthetics (player experience)

* **Narrative / Discovery** → exploring who I, CONSCIOUSNESS, and ENTITY are
* **Uneasiness / Sensation** → experiencing silences, glitches, and disruptions as tension
* **Contemplation** → reflecting on the nature of choices and fragmented identity
* **Expression** → shaping I’s path through moral decisions
* **Submission** → accepting ENTITY’s autonomy and partial loss of control

## Implementation Status

**✓ Implemented (facts)**

* Working engine in Python with 6 demo scenes.
* ENTITY runs on a Temporal Controller with probabilistic gating, emotions, and silence as valid action.
* Classic scenes and metronarrative scenes (on-beat, off-beat, rest, break) already functional.

**~ Observed (hypotheses)**

* Silence from ENTITY creates tension and anticipation.
* Beat synchronization influences perception of stability vs. disruption.
* Attribute shifts make players perceive “identity convergence.”

**○ To Validate (speculation)**

* Whether “music as narrative space” is universally perceived as intended.
* If metronarrative can be generalized into a broader design framework.
