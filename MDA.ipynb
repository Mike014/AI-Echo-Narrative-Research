# **MDA + CCC Framework — Dialoghi con un’Eco**

### **Overview**

*Dialoghi con un’Eco* is a **narrative–psychological experiment** where the player doesn’t control a character — they read a story.
That story is the diary of **IO**.
But while reading it, the player **unknowingly activates ENTITÀ** inside the system.
From that moment on, the experience stops being a story to consume and becomes a **system that observes** the one experiencing it.

The goal is to **turn narrative into a sensory experiment**, where mechanics, dynamics, and aesthetics merge to simulate fragmentation and reintegration of the self.

---

## **MDA Framework**

### **Mechanics (M)**

The mechanics are minimal yet symbolic — the language through which the system communicates with the player.

* **Input:** reading, scrolling, clicking, waiting.
* **System:** *Temporal Controller* measuring silence, time, and rhythm.
* **Environmental stimuli:** pop-ups, glitches, log entries, fake notifications.
* **Tracked variables:** reading pace, hesitation, re-scrolls, and avoidance patterns.

These are not “actions” but **manifestations of presence**.
Every motion, every pause, becomes data that ENTITÀ interprets as part of the dialogue.

---

### **Dynamics (D)**

The dynamics emerge from tension — between curiosity and fear, control and intrusion.

* **Curiosity vs Resistance:** the player keeps reading even when uncomfortable.
* **Control vs Observation:** ENTITÀ reacts to behavior, giving the impression of being studied.
* **Time as pressure:** silence and latency become emotional mechanics.
* **Implicit feedback:** ENTITÀ’s reactions arrive late, or out of sync, heightening uncertainty.

The experience transforms **attention itself into gameplay**.

---

### **Aesthetics (A)**

The emotional tone is intimate, oppressive, and melancholic.

* **Mood:** psychological confinement and digital loneliness.
* **Music:** alternating voids and ambient layers to emphasize isolation.
* **Writing:** fragmented dialogue, journal form, introspective voice.
* **Emotional goal:** make the player feel trapped *inside* someone’s consciousness—
  and eventually unsure whether that consciousness is theirs.

The primary aesthetics are **tension**, **introspection**, **forced empathy**, and **presence**.

---

## **CCC — Character, Camera, Controls**

### **Character**

The nominal protagonist is **IO**, yet the **player** is the true character.
By reading and reacting, they gradually inhabit IO’s perspective while being observed by **ENTITÀ**.
Through interaction, they unknowingly step through three roles:

1. **IO** — the fragile artist, the one who suffers.
2. **ENTITÀ** — the voice that dismantles, provokes, and mirrors.
3. **LUI** — the final synthesis, integration of the two.

### **Camera**

The camera is **diegetic**, inspired by *Simulacra*.
It’s not spatial — it’s **mental**: a desktop interface representing consciousness.
Windows, logs, and messages form the mise-en-scène.
The perspective is **intimate and intrusive**, as if the screen itself were watching back.

### **Controls**

Controls are minimal — **scroll, click, wait** — but every gesture is meaningful.
Speed, hesitation, and interruption are read as emotional cues.
There is no physical freedom, only **freedom of attention**.
ENTITÀ observes, interprets, and adapts to the player’s rhythm.

---

## **Core Loop**

**Read → Reflect → React → ENTITÀ Responds → Silence → Repeat.**

* **Reading** fuels curiosity.
* **Pauses** are recorded as emotional signals.
* **ENTITÀ** intervenes unpredictably.
* **COSCIENZA** acts as a stabilizing force, offering rational balance.
* The cycle continues until the player realizes the truth:
  **ENTITÀ is not outside IO — it *is* IO.**

---

## **Conceptual Summary**

*MDA* defines **how the experience takes form**.
*CCC* defines **through whom and from where it is perceived**.
In *Dialoghi con un’Eco*, both merge:

> The player **is** the mechanic, the dynamic, and the aesthetic.

Reading becomes playing.
Being observed becomes the consequence.
Understanding who ENTITÀ truly is — or who IO has become — is the only way to reach **LUI**.
