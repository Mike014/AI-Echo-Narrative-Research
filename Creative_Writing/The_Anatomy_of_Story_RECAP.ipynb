# Chapter 1 — Story Space, Story Time

## Central Thesis

* A **great story** is a living organism: it communicates a **dramatic code** (human growth/transform­ation) through characters, actions, and choices.
* Audiences don’t see the “real world” but a **story world**: life condensed and heightened to reveal how life works.

## Problem to Solve

* Classical terms (climax, rising action…) and **three-act structure** are too abstract/mechanical → they produce **episodic**, generic stories.
* The “easy process”: derivative premise, overstuffed hero, weak opponents, theme avoided or lectured, dialogue that only pushes plot.

## The Dramatic Code (what all stories do)

* **Desire → action + learning → decisions → change** (or a justified non-change).
* Forms emphasize different phases: myth (vast action), detective/multiperspective (learning), film (intense micro-change), short story (single insight), novel (individual↔society or inner process), TV (simultaneous change in a mini-society).
* Drama is a **code of maturation**: breaking habits/ghosts of the past toward a better self.

## The Story Body (interdependent subsystems)

* **Theme / moral argument** = brain; **characters** = heart/circulation; **revelations** = nervous system; **structure** = skeleton; **scenes** = skin.
* Every part works **only in relation** to the others (mutually defining webs).

## Story Movement (natural patterns)

* **Linear** (single hero, strong desire, cause–effect).
* **Meandering** (wandering journey, varied social encounters).
* **Spiral** (obsessive returns to an event/memory, ever deeper).
* **Branching** (choral fresco/systems and subcultures).
* **Explosive (apparent)**: felt simultaneity via **cross-cut**; compares elements and maps the world.

## The Narrative Game: teller and listener

* The storyteller **selects/connects** high-intensity moments and **hides/reveals** information. The audience both **feels** (relives) and **thinks** (solves the puzzle). When there’s nothing left to figure out, the story ends.

## The Process that Works (internal, organic, interconnected, original)

1. **Premise (1 sentence)**: the story’s essence.
2. **7 Structural Steps (the DNA)**: Weakness/Need, Desire, Opponent, Plan, Battle, Self-Revelation, New Equilibrium.
3. **Character**: born from the idea and **related** to each other (each advances the hero’s evolution).
4. **Theme / Moral Argument**: implicit in structure, not preached.
5. **Story World**: extension/pressure on the hero; visualizes growth.
6. **Symbol Web**: compressed meanings amplifying people/world/plot.
7. **Plot (22 steps)**: events connected beneath the surface → an ending that’s surprising yet necessary.
8. **Scene Weave**: full scene list with plots/themes interlaced.
9. **Scenes & Symphonic Dialogue**: each scene advances the evolution; dialogue operates on multiple levels (objective, subtext, values).

## Operational Takeaways

* Build **from the inside out**: desire and need generate structure, world, symbols, and scenes.
* Aim for **organic unity**: many techniques, invisible on the surface; the audience perceives only **life unfolding** and a **necessary change**.

# The Simplest Possible Cheat Sheet: Premise + 7 Steps

## Premise (1 sentence)

Pattern:
**[Protagonist] wants [concrete goal] but [opposing force]; so [main plan/action]. This leads to [self-discovery] and [new state].**

Neutral example:
“A nervous violinist wants to join an orchestra but her tremor sabotages her; she practices secretly in the subway, discovers fear is control not fate, and performs in public without hiding.”

## The 7 Steps (the “DNA”)

1. **Weakness / Need**
   What’s missing (psychological/moral), and what must be learned to become whole?
   Question: **What’s the flaw, and what must change in them?**

2. **Desire**
   External, clear, measurable objective.
   Question: **Precisely what do they want to achieve?**

3. **Opponent**
   Who/what wants the same thing or can truly block it?
   Question: **Whose desire clashes with theirs?**

4. **Plan**
   Concrete strategy to reach the goal.
   Question: **What course of action will they follow?**

5. **Battle**
   The decisive confrontation with maximum stakes.
   Question: **What’s the crucial trial, and what’s really at risk?**

6. **Self-Revelation**
   A truth about self/the world that transforms behavior.
   Question: **What do they finally realize right there?**

7. **New Equilibrium**
   Life after the change (or why it doesn’t change).
   Question: **What’s their world like now, concretely?**
