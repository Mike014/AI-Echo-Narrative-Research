## **1. Clean system logs**

```bash
sudo journalctl --vacuum-size=100M
```

> Reduces systemd logs, keeping only the latest 100 MB.

---

### **2. Remove orphaned or unused packages**

```bash
sudo apt autoremove --purge -y
```

> Deletes unnecessary dependencies and their config files to free space.

---

### **3. Clear APT cache**

```bash
sudo apt clean
```

> Removes all cached `.deb` files from `/var/cache/apt/archives`.

---

### **4. Fix broken packages**

```bash
sudo apt -f install
```

> Repairs broken dependencies or incomplete package installations.

---

### **5. Configure uncompleted packages**

```bash
sudo dpkg --configure -a
```

> Completes the configuration of any packages left half-installed.

---

### **6. Update and upgrade the entire system**

```bash
sudo apt update && sudo apt full-upgrade -y
```

> Updates package lists and upgrades all installed software (preferred on Kali).

---

### **7. (Optional) Check system status**

```bash
uname -r      # shows the running kernel version
df -h         # displays free disk space
neofetch      # overview of system info
```
