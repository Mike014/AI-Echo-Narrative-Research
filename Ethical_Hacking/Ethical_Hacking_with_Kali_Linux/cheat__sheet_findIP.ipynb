### 1) Show IPv4 addresses (interface + IP)

```bash
ip -4 addr show
```

### 2) Show default route / gateway (to determine network range)

```bash
ip route show default
```

### 3) Show neighbor (ARP) table / discovered LAN hosts

```bash
ip neigh show
```

### 4) Ping a specific device (example IP `192.0.2.12`, 4 packets)

```bash
ping -c 4 192.0.2.12
```

### 5) Quick network discovery / discovery-only (non-invasive)

```bash
sudo nmap -sn 192.0.2.0/24
```

(Replace `192.0.2.0/24` with your actual network if different.)

### 6) Simple HTTP test (create folder + static page, then serve it)

```bash
mkdir -p ~/quick_http
echo '<!doctype html><html><body><h1>Hello from Kali</h1></body></html>' > ~/quick_http/index.html
cd ~/quick_http
python3 -m http.server 8000 --bind 0.0.0.0
```

Then open on your phone’s browser:

```
http://192.0.2.10:8000
```

(where `192.0.2.10` is the **example** IP of Kali — replace with your Kali IP when testing).

---

### 7) (Optional) Simple ping sweep if you don’t use nmap

```bash
for ip in 192.0.2.{1..254}; do ping -c 1 -W 0.3 $ip >/dev/null && echo "$ip responds"; done
```
