# 0) Premise

* **The designer creates experiences, not objects.** The artifact (rules, systems, UI, assets, code) is the means; the designer is judged on the experience it induces.
* **We do not directly control experiences:** we shape the artifact to **induce** desired states.

# 1) Lens #1 — The Lens of Emotion

**Guiding Questions**
* What emotions do I want to evoke? Why?
* What emotions do I currently see in players? Why?
* How can I bridge the gap between the desired and the actual? (By changing rules, feedback, pacing, aesthetics, challenge?)

# 2) Lens #2 — The Lens of Essential Experience

**Principle**
* The goal isn't to replicate reality, but to **capture its essence** (the defining elements) and discard accidental details.
* Example "snowball fight": the **cold** could emerge from visuals (condensation, shivering), audio (wind), or even **rules** (bare hands = better throws but risk of "freezing").

**Guiding Questions**
* What experience do I want the player to have?
* What is **essential** to that experience?
* How does my game **capture** that essence?
* If there's a gap between the desired and actual experience, **I change the game**.

> Note: A designed experience can feel **more "real" than everyday life**.

# 3) Methods for Designing Experience

* **Psychology:**
  * *Behaviorism* (measurable, repeatable): A/B testing, telemetry, controlled experiments.
  * *Phenomenology* (subjective): introspection, usage diaries, qualitative interviews.
    → The designer can and should use **both**.
* **Cultural Anthropology:** "Field work" with players; immerse, observe, interview, understand meanings from the inside.
* **Cross-Disciplinary Design:** Build bridges to music, architecture, film, choreography, UI/UX... to import **empirical rules** and patterns.

# 4) Games as Problem Solving (and the "Magic Circle")

* A game is (also) **a problem-solving activity, approached with a playful attitude**.
* The brain constructs a **closed micro-reality** with goals and rules: this is the origin of the **"magic circle"** and endogenous value (what matters in the game because it matters for winning/progressing).
* Even atypical cases (gambling, *War*) activate attempts at control until the inevitability of chance is learned.

# 5) The Elemental Tetrad (All are Indispensable)

1.  **Mechanics** — Rules/procedures/goals; they distinguish games from linear media.
    * Must be supported by technology, made clear by aesthetics, and meaningful in the story.
2.  **Story** — A sequence of events (linear/branching/emergent); choose mechanics that **reinforce it** and allow for narrative emergence; aesthetics/technology support its tone and pace.
3.  **Aesthetics** — Look, sound, feel; the most direct channel to the experience; must be **orchestrated** and coherent with mechanics and story.
4.  **Technology** — The medium/platform (paper, engine, hardware); enables constraints and possibilities; it is the **medium** where aesthetics, mechanics, and story occur.

> No single element is "more important": there is a **gradient of visibility** (Technology least visible, Aesthetics most visible), but strength comes from **overall coherence**.

# 6) Artifact vs. Experience

* **Artifact**: Rules, systems, UI, assets, code (what you make).
* **Experience**: Perceptions, emotions, sense of control, meaning, memory (what the player lives).
* If no one plays, only the object exists; **the experience does not exist**.

# 7) Playtest — Application to *Dialoghi con un'Eco: Terminal*

* **Automate functionality** (pytest/Selenium) to free up time for experiential tuning.
* **"Essential but open" rules**: Coherent with the game's philosophy, allowing for player interpretation, enhancing immersion.
* **Explicit experiential hypotheses for each mechanic**:
  * "This rule should induce ___ because ___; we verify it by observing ___ and measuring ___."
* **Verification cycles**: Qualitative observation (phenomenology) + metrics/behaviors (behaviorism).
* **Success criterion**: Reduction of the **Emotions gap** (Lens #1) and alignment with the **essence** (Lens #2).

# 8) Definitions as Thinking Tools

* They serve to **clarify ideas**, not to rigidly fix terms. If a definition doesn't aid design, change it.

## Operational Check-list (Quick)

* [ ] **Lens #1**: List desired vs. observed emotions → define actions to bridge the gap.
* [ ] **Lens #2**: Declare the **essence** (3–5 points) → map how each element of the tetrad conveys it.
* [ ] **Tetrad**: Verify mutual coherence (mechanics, story, aesthetics, technology).
* [ ] **Hypothesis per mechanic** → qualitative + quantitative tests.
* [ ] **Refactor** the artifact guided by the experience (not the other way around).
* [ ] **Re-evaluation**: Is the current experience closer to the essence? If not, iterate.
